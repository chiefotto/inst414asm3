```python
import pandas as pd
import json
import numpy as np
import math
import sklearn
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics.pairwise import cosine_distances
from sklearn.metrics import DistanceMetric
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import jaccard_score

df = pd.read_csv('Fifa 23 Players Data.csv')

df = df.set_index('Full Name')
df
```

```python
# normilization and player attribvute columns established (striker attributes)

striker_attr = ['Height(in cm)', 'Pace Total','Shooting Total','Weak Foot Rating']

striker_df=(df[striker_attr]-df[striker_attr].min())/(df[striker_attr].max()-df[striker_attr].min())

striker_df
```

```python

ronaldo = 'C. Ronaldo dos Santos Aveiro'
ronaldo_stats = striker_df.loc[ronaldo]

gen_stats = striker_df[striker_df.index!=ronaldo]


ronaldo_stats = ronaldo_stats.values.reshape(1,-1)

ronaldo_stats

striker_df['Cosine_Similarity'] = striker_df.apply(lambda row: cosine_similarity(row.values.reshape(1, -1), ronaldo_stats)[0, 0], axis=1)



striker_df_sorted = striker_df.sort_values(by='Cosine_Similarity', ascending=False)

top = striker_df_sorted.head(10)
top
```

```python
passing_attr = ['Passing Total','Crossing','Curve','LongPassing','Short Passing']
passing_df = (df[passing_attr]-df[passing_attr].min())/(df[passing_attr].max()-df[passing_attr].min())

modric = 'Luka Modrić'
modric_stats = passing_df.loc[modric]
modric_stats = modric_stats.values.reshape(1,-1)



def jaccard_similarity(p1,p2):
    intersection = len(set(p1)& set(p2))
    union = len(set(p1)| set(p2))
    return intersection / union



passing_df['Jaccard_Similarity'] = passing_df.apply(lambda row: jaccard_similarity(str(row.values), str(modric_stats)), axis=1)


passing_df = passing_df.sort_values(by='Jaccard_Similarity', ascending=False)

top_10_closest = passing_df.head(50)

top_10_closest

```
