```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.metrics.pairwise import cosine_similarity


df = pd.read_csv('Fifa 23 Players Data.csv')

df = df.set_index('Full Name')
df
```

```python
#  and player attribvute columns established (striker attributes)

striker_attr = ['Height(in cm)', 'Pace Total','Shooting Total','Weak Foot Rating']

# striker_df=(df[striker_attr]-df[striker_attr].min())/(df[striker_attr].max()-df[striker_attr].min())

striker_df = df[striker_attr]

striker_df
```

```python
### Gathering Ronaldos Stats from the filtered dataframe
ronaldo = 'C. Ronaldo dos Santos Aveiro'
ronaldo_stats = striker_df.loc[ronaldo]

### all other player stats
gen_stats = striker_df[striker_df.index!=ronaldo]

### reshaped array
ronaldo_stats = ronaldo_stats.values.reshape(1,-1)


### create a new column with each players cosine similarity
striker_df['Cosine_Similarity'] = striker_df.apply(lambda row: cosine_similarity(row.values.reshape(1, -1), ronaldo_stats)[0, 0], axis=1)


### sort the scores
striker_df_sorted = striker_df.sort_values(by='Cosine_Similarity', ascending=False)
top = striker_df_sorted.head(11)
top
```

```python
passing_attr = ['Crossing','Curve','LongPassing','Short Passing']
# passing_df = (df[passing_attr]-df[passing_attr].min())/(df[passing_attr].max()-df[passing_attr].min())
passing_df = df[passing_attr]


modric = 'Luka Modrić'
modric_stats = passing_df.loc[modric]
modric_stats = modric_stats.values.reshape(1,-1)


def jaccard_similarity(p1,p2):
    intersection = len(set(p1)& set(p2))
    union = len(set(p1)| set(p2))
    return intersection / union

passing_df['Jaccard_Similarity'] = passing_df.apply(lambda row: jaccard_similarity(str(row.values), str(modric_stats)), axis=1)


passing_df = passing_df.sort_values(by='Jaccard_Similarity', ascending=False)

top_10_closest = passing_df.head(15)

top_10_closest

```

```python
messi_attr = ['Sprint Speed','Short Passing','Dribbling','BallControl','Acceleration','Balance']
# messi_df = (df[messi_attr]-df[messi_attr].min())/(df[messi_attr].max()-df[messi_attr].min())
messi_df = df[messi_attr]
messi = 'Lionel Messi'

messi_stats = messi_df.loc[messi]
messi_stats = messi_stats.values.reshape(1,-1)

messi_df['Eucledian_Distance'] = euclidean_distances((messi_df.values), (messi_stats))

messi_df = messi_df.sort_values(by='Eucledian_Distance', ascending=True)

top_10_closest = messi_df.head(11)

top_10_closest
```
